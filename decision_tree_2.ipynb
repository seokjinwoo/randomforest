````
cleanup_nums = {"health_insurance": {1:1, 2:1, 3:0}}
df = df.replace(cleanup_nums)
df.head()
````

````
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree 
from sklearn.model_selection import train_test_split
````

````
x_list = ['Schooling','Health_status','selfesteem','satisfaction','aggression','satisfaction','mental','depression','community','stigma','future_hope']
x_list = ['Schooling','Health_status','selfesteem','depression','stigma','future_hope']
x_list 
````

````


X = df[x_list ]
y = df[['health_insurance']]


X_train,X_test,y_train,y_test = train_test_split(X,y, random_state = 0)
X_train.head()
````

````

clf = DecisionTreeClassifier(max_depth=3)


clf.fit(X_train,y_train)
````

````
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
````

````
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
````
